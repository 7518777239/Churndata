```python
# Calculate percentage distribution
counts = df.groupby(['SeniorCitizen', 'Churn']).size().unstack(fill_value=0)
percentages = counts.div(counts.sum(axis=1), axis=0) * 100

# Plot stacked bar chart
percentages.plot(kind='bar', stacked=True, figsize=(4,4), colormap='viridis')

# Add percentage labels
for i, (index, row) in enumerate(percentages.iterrows()):
    cumulative = 0
    for churn_status, value in row.items():
        plt.text(i, cumulative + value / 2, f'{value:.1f}%', ha='center', va='center', fontsize=10, color='white')
        cumulative += value

plt.xlabel("SeniorCitizen")
plt.ylabel("Percentage")
plt.title("SeniorCitizen by Churn (%)")
plt.legend(title='Churn')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("Customer Churn.csv")
df
```

```python
plt.figure(figsize=(9, 4))
sns.histplot(x="tenure", data=df, bins=30, hue="Churn")
plt.show()
```

## People who have used oure services for a long time have stayed and people who have used oure services #1 or 2 months have churned.

## Contract

```python
plt.figure(figsize=(4,4))
ax= sns.countplot(x= "Contract", data = df, hue = "Churn")
ax.bar_label(ax.containers[0])
plt.title("Count of customer by contract")
plt.show()
```

## People who have month to month contract are to churn then from those who have 1 or 2 years or contract.

```python
df.columns.values
```

```python
# List of columns
columns = ['PhoneService', 'MultipleLines', 'InternetService',
           'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
           'TechSupport', 'StreamingTV', 'StreamingMovies']

# Set up the subplot grid
fig, axes = plt.subplots(3, 3, figsize=(12
                                        ,9))
axes = axes.flatten()  # Flatten the axes array for easy iteration

# Plot each countplot
for i, col in enumerate(columns):
    sns.countplot(x=col, data=df, ax=axes[i], palette='Set2',hue="Churn")
    axes[i].set_title(f'Countplot of {col}')
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Count')
    axes[i].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

## The subplots illustrate the distribution of customer churn (Yes or No) across various services. Customers without additional services like OnlineSecurity, TechSupport, and OnlineBackup show a higher churn rate. Conversely, customers with active services like PhoneService and InternetService tend to have lower churn. Overall, lacking value-added services is linked to increased churn.
